**Q1)**	Give 4 examples of **fetching Rows in a Pandas DataFrame**. (Read the given csv file into Python using Pandas)

```python
import pandas as pd
```

#read the file into python using pandas


```python
df = pd.read_csv("Sleep_Efficiency.csv")
```

```python
df  # loading the data
```

```python
df.iloc[4,:] # selecting the 5th row
```

```python
df.iloc[[4],:] # selecting the 5th row using a list to have a proper formart
```

```python
df.iloc[[1,2],:] # selecting the 2nd and 3rd row (multiple rows)
```

```python
df.iloc[1:5,:]  # selecting the 2nd to 4th row using range
# this gives the rows from 2 to 4, but not including the 5th row.
```

```python
df.loc[1:5,:] # selecting the 2nd to 5th row using loc method and range
# this gives the rows from 1 to 5 inclusive of 5.
```

**Q3)** Give some examples and Write few points to explain the **difference between loc and .iloc** attributes in a Pandas DataFrame. (Read the given csv file into Python using Pandas)

difference between iloc and loc :
loc:-
1. we can fetch the rows or columns including the end values
2. works with lables.
iloc:-
1. iloc fetches the rows or columns excluding the end value
2. integer based indexing and numerical positions.

eg [0:5] - range of values from 0 to 5
loc fetches all the values from 1 to 5 including 5 , whereas iloc fetches data from 0 till 4 excluding 5th row.


```python
df_1.iloc[0:4,:] # fetching data from 1st to 4th column using range not including 5th column
```

```python
df_1.loc[0:4] # fetching data from 1st to 4th column using loc method and range including 5th column
```
